```python
import matplotlib.pyplot as plt
import struct
import numpy as np
%matplotlib inline
```

```python
f = open("./0000005-disp.pfm",'rb')
```

```python
hdr1 = f.readline().strip()  # Pf grayscale,  PF color
hdr1
```

```python
hdr2 = f.readline().strip().split()  # resolution
x = int(hdr2[0])
y = int(hdr2[1])
x,y
```

```python
hdr3 = f.readline() # +1 big endian,  -1 little endian
hdr3
```

```python
pixels = []
while True:
    i = f.read(4)
    if not i: break
    i = struct.unpack(">f",i)
    pixels.append(i)
f.close()
```

```python
plt.plot(pixels)
plt.show()
```

```python
img = np.asarray(pixels, dtype=np.float32,order='C' )
```
